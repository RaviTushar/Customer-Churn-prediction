```python
# ==============================
# 6. CONTRACT & BILLING ANALYSIS
# ==============================

import pandas as pd

# Load dataset
df = pd.read_csv("/kaggle/input/datasets/blastchar/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv")

# ==============================
# 1. Contract Type Distribution
# ==============================
print("\n--- Contract Type Distribution ---")
print(df['Contract'].value_counts())

print("\nPercentage:")
print(df['Contract'].value_counts(normalize=True) * 100)

# ==============================
# 2. Paperless Billing
# ==============================
print("\n--- Paperless Billing ---")
print(df['PaperlessBilling'].value_counts())

print("\nPercentage:")
print(df['PaperlessBilling'].value_counts(normalize=True) * 100)

# ==============================
# 3. Payment Method Distribution
# ==============================
print("\n--- Payment Method Distribution ---")
print(df['PaymentMethod'].value_counts())

print("\nPercentage:")
print(df['PaymentMethod'].value_counts(normalize=True) * 100)

# ==============================
# 4. Monthly Charges Summary
# ==============================
print("\n--- Monthly Charges Statistics ---")
print(df['MonthlyCharges'].describe())

# ==============================
# 5. Total Charges Summary
# ==============================
print("\n--- Total Charges Preview (converted if needed) ---")

# Convert TotalCharges to numeric (important fix in dataset)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

print(df['TotalCharges'].describe())

# ==============================
# 6. Contract vs Average Monthly Charges
# ==============================
print("\n--- Average Monthly Charges by Contract Type ---")
print(df.groupby('Contract')['MonthlyCharges'].mean())

# ==============================
# 7. Contract vs Total Charges
# ==============================
print("\n--- Average Total Charges by Contract Type ---")
print(df.groupby('Contract')['TotalCharges'].mean())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/kaggle/input/datasets/blastchar/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv")

plt.figure(figsize=(6,4))
df['Contract'].value_counts().plot(kind='bar')

plt.title('Contract Type Distribution')
plt.xlabel('Contract Type')
plt.ylabel('Number of Customers')
plt.show()
```

```python
plt.figure(figsize=(7,4))
df['PaymentMethod'].value_counts().plot(kind='bar')

plt.title('Payment Method Distribution')
plt.xlabel('Payment Method')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

plt.figure(figsize=(6,4))
df.groupby('Contract')['TotalCharges'].mean().plot(kind='bar')

plt.title('Average Total Charges by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Average Total Charges')
plt.show()
```
