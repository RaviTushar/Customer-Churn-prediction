```python
# ==============================
# 4. CUSTOMER DEMOGRAPHICS ANALYSIS (TABLE OUTPUT ONLY)
# ==============================

import pandas as pd

# Load dataset
df = pd.read_csv("/kaggle/input/datasets/blastchar/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv")

# ==============================
# 1. Gender Distribution
# ==============================
print("\n--- Gender Distribution ---")
print(df['gender'].value_counts())
print("\nPercentage:")
print(df['gender'].value_counts(normalize=True) * 100)

# ==============================
# 2. Senior Citizen Distribution
# ==============================
print("\n--- Senior Citizen Distribution ---")
print(df['SeniorCitizen'].value_counts())
print("\nPercentage:")
print(df['SeniorCitizen'].value_counts(normalize=True) * 100)

# ==============================
# 3. Partner Status Distribution
# ==============================
print("\n--- Partner Distribution ---")
print(df['Partner'].value_counts())
print("\nPercentage:")
print(df['Partner'].value_counts(normalize=True) * 100)

# ==============================
# 4. Dependents Distribution
# ==============================
print("\n--- Dependents Distribution ---")
print(df['Dependents'].value_counts())
print("\nPercentage:")
print(df['Dependents'].value_counts(normalize=True) * 100)

# ==============================
# 5. Tenure Summary Statistics
# ==============================
print("\n--- Tenure Statistics ---")
print(df['tenure'].describe())

# ==============================
# 6. Monthly Charges Summary Statistics
# ==============================
print("\n--- Monthly Charges Statistics ---")
print(df['MonthlyCharges'].describe())

# ==============================
# 7. Basic Demographic Summary
# ==============================
print("\n--- Overall Dataset Size ---")
print("Total Customers:", df.shape[0])
print("Total Features:", df.shape[1])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/kaggle/input/datasets/blastchar/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv")

fig, ax = plt.subplots(1, 2, figsize=(12,5))

# Gender
df['gender'].value_counts().plot(kind='bar', ax=ax[0])
ax[0].set_title('Gender Distribution')
ax[0].set_xlabel('Gender')
ax[0].set_ylabel('Count')

# Senior Citizen
df['SeniorCitizen'].value_counts().plot(kind='bar', ax=ax[1])
ax[1].set_title('Senior Citizen Distribution')
ax[1].set_xlabel('Senior Citizen (0 = No, 1 = Yes)')
ax[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,5))

# Partner
df['Partner'].value_counts().plot(kind='bar', ax=ax[0])
ax[0].set_title('Partner Status Distribution')
ax[0].set_xlabel('Partner')
ax[0].set_ylabel('Count')

# Dependents
df['Dependents'].value_counts().plot(kind='bar', ax=ax[1])
ax[1].set_title('Dependents Distribution')
ax[1].set_xlabel('Dependents')
ax[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
