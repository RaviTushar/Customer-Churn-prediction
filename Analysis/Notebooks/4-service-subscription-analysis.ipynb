```python
# ==============================
# 5. SERVICE SUBSCRIPTION ANALYSIS (TABLE OUTPUT ONLY)
# ==============================

import pandas as pd

# Load dataset
df = pd.read_csv("/kaggle/input/datasets/blastchar/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv")

# ==============================
# 1. Phone Service
# ==============================
print("\n--- Phone Service ---")
print(df['PhoneService'].value_counts())

print("\nPercentage:")
print(df['PhoneService'].value_counts(normalize=True) * 100)

# ==============================
# 2. Internet Service
# ==============================
print("\n--- Internet Service ---")
print(df['InternetService'].value_counts())

print("\nPercentage:")
print(df['InternetService'].value_counts(normalize=True) * 100)

# ==============================
# 3. Online Security
# ==============================
print("\n--- Online Security ---")
print(df['OnlineSecurity'].value_counts())

print("\nPercentage:")
print(df['OnlineSecurity'].value_counts(normalize=True) * 100)

# ==============================
# 4. Tech Support
# ==============================
print("\n--- Tech Support ---")
print(df['TechSupport'].value_counts())

print("\nPercentage:")
print(df['TechSupport'].value_counts(normalize=True) * 100)

# ==============================
# 5. Streaming TV
# ==============================
print("\n--- Streaming TV ---")
print(df['StreamingTV'].value_counts())

print("\nPercentage:")
print(df['StreamingTV'].value_counts(normalize=True) * 100)

# ==============================
# 6. Streaming Movies
# ==============================
print("\n--- Streaming Movies ---")
print(df['StreamingMovies'].value_counts())

print("\nPercentage:")
print(df['StreamingMovies'].value_counts(normalize=True) * 100)

# ==============================
# 7. Quick Summary Counts
# ==============================
print("\n--- Dataset Summary ---")
print("Total Customers:", df.shape[0])
print("Total Features:", df.shape[1])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/kaggle/input/datasets/blastchar/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv")

plt.figure(figsize=(6,4))
df['InternetService'].value_counts().plot(kind='bar')

plt.title('Internet Service Distribution')
plt.xlabel('Internet Service Type')
plt.ylabel('Number of Customers')
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,5))

df['OnlineSecurity'].value_counts().plot(kind='bar', ax=ax[0])
ax[0].set_title('Online Security Usage')
ax[0].set_xlabel('Status')
ax[0].set_ylabel('Count')

df['TechSupport'].value_counts().plot(kind='bar', ax=ax[1])
ax[1].set_title('Tech Support Usage')
ax[1].set_xlabel('Status')
ax[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,5))

df['StreamingTV'].value_counts().plot(kind='bar', ax=ax[0])
ax[0].set_title('Streaming TV Usage')
ax[0].set_xlabel('Status')
ax[0].set_ylabel('Count')

df['StreamingMovies'].value_counts().plot(kind='bar', ax=ax[1])
ax[1].set_title('Streaming Movies Usage')
ax[1].set_xlabel('Status')
ax[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
