```python
# ==============================
# 7. CHURN ANALYSIS
# ==============================

import pandas as pd

# Load dataset
df = pd.read_csv("/kaggle/input/datasets/blastchar/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv")

# ==============================
# 1. Overall Churn Distribution
# ==============================
print("\n--- Churn Distribution ---")
print(df['Churn'].value_counts())

print("\nPercentage:")
print(df['Churn'].value_counts(normalize=True) * 100)

# ==============================
# 2. Churn vs Contract Type
# ==============================
print("\n--- Churn vs Contract Type (Counts) ---")
print(pd.crosstab(df['Contract'], df['Churn']))

print("\n--- Churn vs Contract Type (Row Percentage) ---")
print(pd.crosstab(df['Contract'], df['Churn'], normalize='index') * 100)

# ==============================
# 3. Churn vs Payment Method
# ==============================
print("\n--- Churn vs Payment Method (Counts) ---")
print(pd.crosstab(df['PaymentMethod'], df['Churn']))

print("\n--- Churn vs Payment Method (Row Percentage) ---")
print(pd.crosstab(df['PaymentMethod'], df['Churn'], normalize='index') * 100)

# ==============================
# 4. Churn vs Internet Service
# ==============================
print("\n--- Churn vs Internet Service (Counts) ---")
print(pd.crosstab(df['InternetService'], df['Churn']))

print("\n--- Churn vs Internet Service (Row Percentage) ---")
print(pd.crosstab(df['InternetService'], df['Churn'], normalize='index') * 100)

# ==============================
# 5. Churn vs Tenure (Statistical View)
# ==============================
print("\n--- Tenure Statistics by Churn ---")
print(df.groupby('Churn')['tenure'].describe())

# ==============================
# 6. Churn vs Monthly Charges
# ==============================
print("\n--- Monthly Charges Statistics by Churn ---")
print(df.groupby('Churn')['MonthlyCharges'].describe())

# ==============================
# 7. Overall Summary
# ==============================
print("\n--- Dataset Summary ---")
print("Total Customers:", df.shape[0])
print("Total Features:", df.shape[1])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/kaggle/input/datasets/blastchar/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv")

plt.figure(figsize=(5,4))
df['Churn'].value_counts().plot(kind='bar')

plt.title('Overall Churn Distribution')
plt.xlabel('Churn')
plt.ylabel('Number of Customers')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/kaggle/input/datasets/blastchar/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv")

ct = pd.crosstab(df['Contract'], df['Churn'], normalize='index') * 100

ct.plot(kind='bar', figsize=(6,4))

plt.title('Churn Rate by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Churn Percentage')
plt.legend(title='Churn')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(6,4))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)

plt.title('Monthly Charges vs Churn')
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')
plt.show()
```
