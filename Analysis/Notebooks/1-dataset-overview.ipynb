```python
import pandas as pd

df = pd.read_csv("/kaggle/input/datasets/blastchar/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv")

print("Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns.tolist())

print("\nInfo:")
print(df.info())

print("\nFirst 5 Rows:")
print(df.head())


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Count feature types
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

feature_types = pd.DataFrame({
    'Feature Type': ['Categorical', 'Numerical'],
    'Count': [len(categorical_cols), len(numerical_cols)]
})

# Plot
plt.figure(figsize=(6,4))
sns.barplot(data=feature_types, x='Feature Type', y='Count')

plt.title('Distribution of Feature Types')
plt.xlabel('Feature Type')
plt.ylabel('Number of Features')

for index, value in enumerate(feature_types['Count']):
    plt.text(index, value + 0.1, str(value), ha='center')

plt.tight_layout()
plt.show()
```
